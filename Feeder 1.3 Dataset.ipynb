# CSV File
You will need to define whatever csv file with an object, in this case the object being represent as **ds**, but first, the csv file you are using and have exported into the working directory will need to be accessed into this data file. To do so, you will need to use **pd.read_csv()** and type in the exact name of the csv file into the parentheses.
```python
ds=pd.read_csv("CountyHealthData_2014-2015.csv")
```

# Filtering
Now that the csv file is in, the only information on the file that is needed is the data for North Carolina. In order to show information from only from North Carolina, you will need to filter only that portion of the dataset. This is done by using logical (true or false) conditions with nested square brackets. In these nested square brackets, there will be both inner brackets and outer brackets. Outside of both brackets, you will enter **ds**. In the very inner bracket, you will enter the word **State** in quotations, making sure the s in state is capitalized. The information stored in the outer bracket will be written as **ds == "NC"**. The inner bracket will be placed right between the ds and the double equal signs. The order should be **ds**, then an open bracket and **ds** again with another open bracket (this represents the inner bracket), the word **State** written in quotations, a closed bracket (closing the inner bracket), two equal signs, the letters **NC** written in quotations and another closed bracket (closing the outer bracket). Finally, activate it by pressing both the **ctrl** and **enter** button at the same time.

```python
ds[ds["State"]== "NC"]
```

This is the part of the dataset that will be worked on the most, so a subset name will come in handy instead of having to continually type in the same information. To do so, you will come up with a name for the dataset, this one will be called **NC_subset**, then add an equal sign and lastly take that same code that was typed in for filtering out the information for NC and add **.copy()** at the end of it. Make sure to press **ctrl** and **enter** at the same time after doing so.

```python
NC_subset=ds[ds["State"]== "NC"]
```

Now that we have saved all of the information that is provided for North Carolina, not all of the columns will be needed. In order to only show the columns needed, you will type in the saved dataset **NC_subset** and type in **.loc** afterwards to only include the columns needed. You will then need to add an inner and outer bracket once again. The outer bracket will contain a colon(**:**) with a comma after the colon and then add an open bracket which will present the inner bracket. Into the inner bracket add the words **"State"** with a comma afterwards, then **"County"** with a comma afterwards, then **"Physical inactivity"** with a comma afterwards, then lastly **"Access to exercise opportunities"** with two closing brackets afterwards. Make sure to include the quotations for each word in too. Lastly click both the **ctrl** and **enter** button at the same time.

```python
NC_subset.loc[:,["State","County","Physical inactivity","Access to exercise opportunities"]]
```

Now save this condensed data as its own data by coming up with a name, this one will be called **NC_data**, and add an equal sign and the rest of the code used to create that condensed data.

```python
NC_data=NC_subset.loc[:,["State","County","Physical inactivity","Access to exercise opportunities"]] 
```

# Saving as CSV File
Now to save this information as a csv file, you will need to take **NC_data** and add **.to_csv** so that it will be saved as a csv file. Then add a parentheses and inside of the parantheses, name the file and add  quotations around the name. This particular file will be called **NC_data.csv**. To get rid of the numbers in front of information that was pulled out, you will add a comma, then a space and enter **index=False** making sure to capitalize the f in false. Lastly, press **ctrl** and **enter** at the same time.

```python
NC_data.to_csv("NC_data.csv", index=False)
```

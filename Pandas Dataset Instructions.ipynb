# Filtering
## Retrieving NC Data
Now that the csv file is in, the only information on the file that is needed is the data for North Carolina. In order to show information from only from North Carolina, you will need to filter only that portion of the dataset. 

This is done by using logical (true or false) conditions with nested square brackets. In these nested square brackets, there will be both inner brackets and outer brackets. Outside of both brackets, you will enter `ds`. In the very inner bracket, you will enter the word `"State"` in quotations, making sure the s in state is capitalized. The information stored in the outer bracket will be written as `ds == "NC"`. The inner bracket will be placed right between the ds and the double equal signs. 

The order should be:
1. `ds`
2. An open bracket and `ds` again with another open bracket (this represents the inner bracket) 
3. The word `"State"` written in quotations
4. A closed bracket (closing the inner bracket)
5. Two equal signs
6. The letters `NC` written in quotations 
7. Another closed bracket (closing the outer bracket) 
Finally, activate it by pressing both the **ctrl** and **enter** button at the same time.

```python
ds[ds["State"]== "NC"]
```

## Saving NC Data
This is the part of the dataset that will be worked on the most, so a subset name will come in handy instead of having to continually type in the same information. 

To create the new subset:
1. You will come up with a name for the dataset, this one will be called `NC_subset`
2. Add an equal sign
3. Take that same code that was typed in for filtering out the information for NC and add `.copy()` at the end of it. 
Make sure to press **ctrl** and **enter** at the same time after doing so.

```python
NC_subset=ds[ds["State"]== "NC"]
```

## Condensing NC Data
Now that we have saved all of the information that is provided for North Carolina, not all of the columns that have been provided for North Carolina will be needed. 

In order to only show the columns needed: 
1. Type in the saved dataset `NC_subset` and type in **.loc** afterwards to only include the columns needed
2. You will then need to add an inner and outer bracket once again. 
- The outer bracket will contain a colon(**:**) with a comma after the colon and then add an open bracket which will present the inner bracket. 
3. Into the inner bracket add the words `"State"` with a comma afterwards, then `"County"` with a comma afterwards, then `"Physical inactivity"` with a comma afterwards, then lastly `"Access to exercise opportunities"` with two closing brackets afterwards. 
- These will be the specific columns that contain information that is actually needed from North Carolina

Make sure to include the quotations for each word in too. Make sure to click both the **ctrl** and **enter** button at the same time after entering the information.

```python
Nc_Data1=NC_subset.loc[:,["State","Year","County","Physical inactivity","Access to exercise opportunities"]]
```

Now save this condensed data as its own data by coming up with a name, this one will be called `NC_data`, and add an equal sign and the rest of the code used to create that condensed data.

# Saving Information as CSV File
Now to save this information as a csv file, you will need to:
1. Take `NC_data` and add `.to_csv` so that it will be saved as a csv file. 
2. Add a parentheses and inside of the parantheses, name the file and add  quotations around the name. (This particular file will be called `NC_data1.csv`.)

To get rid of the numbers in front of information that was pulled out, you will add a comma, then a space and enter `index=False` making sure to capitalize the f in false. 

Lastly, press **ctrl** and **enter** at the same time.

```python
Nc_Data1.to_csv("NC_data1.csv", index=False)
```
